```python
# ПМ-24-6
# Агеев Денис Юрьевич
# 245086

import pandas as pd

"""Загрузка таблицы из CSV файла."""
def load_table_from_csv(file):
    return pd.read_csv(file)
    
    
"""Загрузка таблицы из Pickle файла."""
def load_table_from_pickle(file):
    return pd.read_pickle(file)
    
"""Сохранение таблицы в CSV файл."""
def save_table_to_csv(df, file):
    df.to_csv(file, index=False)
    
"""Сохранение таблицы в Pickle файл."""
def save_table_to_pickle(df, file):
    df.to_pickle(file)
    
"""Получение новой таблицы по индексам."""
def get_rows_by_index(df, *vals):
    return df[df.iloc[:, 0].isin(vals)]
    
"""Получение типов столбцов."""
def get_column_types(df):
    return df.dtypes.to_dict()
    
"""Задание типов столбцов."""
def set_column_types(df, types_dict):
    for col, dtype in types_dict.items():
        if dtype == 'int':
            df[col] = df[col].astype(int)
        elif dtype == 'float':
            df[col] = df[col].astype(float)
        elif dtype == 'bool':
            df[col] = df[col].astype(bool)
        elif dtype == 'str':
            df[col] = df[col].astype(str)
        elif dtype == 'datetime':
            df[col] = pd.to_datetime(df[col])
            
"""Получение значений из столбца."""
def get_values(df, column=0):
    return df.iloc[:, column].tolist()
    
"""Задание значений для столбца."""
def set_values(df, values, column=0):
    if len(values) != len(df):
        raise ValueError("Длина значений не соответствует количеству строк.")
    df.iloc[:, column] = values
    
"""Вывод таблицы на печать."""
def print_table(df):
    print(df)
    
"""Склеивание двух таблиц."""
def concat_tables(df1, df2):
    return pd.concat([df1, df2], ignore_index=True)
    
"""Разбиение таблицы на две по номеру строки."""
def split_table(df, row_number):
    if row_number < 0 or row_number > len(df):
        raise ValueError("Номер строки выходит за пределы.")
    return df.iloc[:row_number], df.iloc[row_number:]

if __name__ == "__main__":
    csv_file = 'data.csv'  # Укажите путь к вашему CSV файлу
    pickle_file = 'data.pkl'  # Укажите путь к вашему Pickle файлу
    
    try:
        table_csv = load_table_from_csv(csv_file)
        print("Table loaded from CSV:")
        print_table(table_csv)
        
        table_pickle = load_table_from_pickle(pickle_file)
        print("Table loaded from Pickle:")
        print_table(table_pickle)
        
        combined_table = concat_tables(table_csv, table_pickle)
        print("Combined Table:")
        print_table(combined_table)

        set_column_types(combined_table, {'D': 'datetime'})

        values = get_values(combined_table, column=1) 
        print("Values from column B:", values)

        set_values(combined_table, [10.0, 20.0, 30.0], column=1) 

        print("Updated Combined Table:")
        print_table(combined_table)

        table1, table2 = split_table(combined_table, 2)
        
        print("Table 1:")
        print_table(table1)
        
        print("Table 2:")
        print_table(table2)
        
        save_table_to_csv(combined_table, 'updated_data.csv')
        save_table_to_pickle(combined_table, 'updated_data.pkl')

    except Exception as e:
        print(f"Произошла ошибка: {e}")
```
